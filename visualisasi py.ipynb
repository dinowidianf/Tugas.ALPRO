```python
import pandas as pd
import matplotlib.pyplot as plt

# ===== BACA DATA CSV =====
data = pd.read_csv("percobaan.csv")

t = data["t"]
x = data["x"]
y = data["y"]

# ===== GRAFIK 1: LINTASAN PARABOLA =====
plt.figure()
plt.plot(x, y)
plt.xlabel("Jarak Horizontal x (m)")
plt.ylabel("Ketinggian y (m)")
plt.title("Lintasan Gerak Parabola Bola")
plt.grid(True)
plt.show()

# ===== GRAFIK 2: POSISI X TERHADAP WAKTU =====
plt.figure()
plt.plot(t, x)
plt.xlabel("Waktu t (s)")
plt.ylabel("Posisi Horizontal x (m)")
plt.title("Posisi Horizontal terhadap Waktu")
plt.grid(True)
plt.show()

# ===== GRAFIK 3: POSISI Y TERHADAP WAKTU =====
plt.figure()
plt.plot(t, y)
plt.xlabel("Waktu t (s)")
plt.ylabel("Ketinggian y (m)")
plt.title("Ketinggian Bola terhadap Waktu")
plt.grid(True)
plt.show()
```
