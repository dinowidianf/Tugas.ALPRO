# Metode Pencarian Beruntun

```python
def sequential_search(arr, x):
     for i in range(len(arr)):
         if arr[i] == x:
             return i
     return -1
 
data_fisika = [10, 50, 30, 70, 80, 20]
target = 30

hasil = sequential_search(data_fisika, target)

if hasil !=-1:
     print(f"Data ditemukan pada indeks: {hasil}")
else:
    print("Data tidak ditemukan.")
```

# Metode Pencarian Biner

```python
def binary_search(arr, x):
    low = 0
    high = len(arr)- 1
    
    while low <= high:
        mid = low + (high- low) // 2
         
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid- 1
            
    return-1

data_energi = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
target = 23

hasil = binary_search(data_energi, target)

if hasil !=-1:
    print(f"Energi ditemukan pada indeks: {hasil}")
else:
    print("Energi tidak ditemukan.")

```

# Pencarian dalam Tabel Massa Jenis

```python
class Element:
    def __init__(self, z, density, name):
        self.z = z
        self.density = density
        self.name = name
        
def find_element_binary(table, target_z):
    low = 0
    high = len(table)- 1
    
    while low <= high:
        mid = low + (high- low) // 2
        current_element = table[mid]
        
        if current_element.z == target_z:
            return mid
        elif current_element.z < target_z:
            low = mid + 1
        else:
            high = mid- 1
            
    return-1

periodic_table = [
    Element(1, 0.000089, "Hydrogen"),
    Element(2, 0.000178, "Helium"),
    Element(6, 2.26, "Carbon"),
    Element(13, 2.70, "Aluminum"),
    Element(14, 2.33, "Silicon"),
    Element(26, 7.87, "Iron"),
    Element(29, 8.96, "Copper"),
    Element(79, 19.32, "Gold")
]

search_z = 79 # Mencari Gold (Emas)
index = find_element_binary(periodic_table, search_z)

print("---Hasil Pencarian---")
if index !=-1:
    el = periodic_table[index]
    print(f"Unsur: {el.name}")
    print(f"Nomor Atom: {el.z}")
    print(f"Massa Jenis: {el.density} g/cm^3")
else:
    print(f"Unsur dengan Z={search_z} tidak ditemukan.")
```

# Metode Pengurutan Apung (Bubble Sort)

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n- i- 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                
data = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(data)
print("Hasil Bubble Sort:", data)
```

# Metode Pengurutan Seleksi (Selection Sort)

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

data = [64, 34, 25, 12, 22, 11, 90]
selection_sort(data)
print("Hasil Selection Sort:", data)
```

# Metode Pengurutan Sisip (Insertion Sort)

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i- 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j-= 1
        arr[j + 1] = key
        
data = [64, 34, 25, 12, 22, 11, 90]
insertion_sort(data)
print("Hasil Insertion Sort:", data)
```

# Metode Pengurutan Gabung (Merge Sort)

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]
        
        merge_sort(L)
        merge_sort(R)
        
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
                k += 1
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
            
data = [64, 34, 25, 12, 22, 11, 90]
merge_sort(data)
print("Hasil Merge Sort:", data)
```
